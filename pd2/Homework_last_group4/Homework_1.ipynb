# Loading Packages

```python
# NetworkX
import networkx as nx

# Math
import numpy as np
import pandas as pd
import random

# Plot
%matplotlib inline
import matplotlib.pyplot as plt
```

# 1. Take an empty graph with only nodes

We create a graph with 50 nodes.

```python
g=nx.empty_graph(50)
nx.draw(g)
plt.title("Empty Graph (50 Nodes)")
```

# 2. Add edges to the graph randomly using Erdős–Rényimodel till you get a completegraph & 3. Calculate the Freemans formula for each step (after adding each edge)

```python
def fill(g, d):
    """fill(g, d)
    
    Fills a graph. Returns Freeman's value for each step. 
    
    Args:
    * g - graph
    * d - degree
    
    Outputs:
    * f - freeman values
    """
    # Create Ret
    f=np.zeros(d*(d-1)//2)

    for i in range(len(f)):
        # Add Edge
        g.add_edge(*random.choice(list(nx.non_edges(g))))

        # Cent Values
        cv=nx.closeness_centrality(g).values()

        # Calculate Formula
        f[i]=(d*max(cv)-sum(cv))/(d-2)
    
    return f
```

```python
f = fill(g, 50)
```

# 4. Plot the results

## 4.1 Full Graph

```python
nx.draw(g)
plt.title('Full Graph After Filling');
```

## Freeman Formula

```python
plt.plot(
    np.arange(len(f)),
    f
)
plt.title("Freeman Values")
plt.xlabel("Step")
plt.ylabel("Value");
```
